```python
df['isFraud']=df['isFraud'].map({0:'No fraud',1:'fraud'})
```

```python
x=df[['type','amount','oldbalanceOrg','newbalanceOrig']]
```

The **iloc** function is a very powerful tool for selecting rows and columns of DataFrames. It is often used in data analysis and machine learning tasks.

```python
y=df.iloc[:,-2] #y is assigned the second-last column of the DataFrame df
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
model.predict([[2,9800,170136,160296]])
```
